```python
#Dependecies

import requests
import pandas as pd
from config import gkey
from config import ykey
import numpy as np
from matplotlib import pyplot as plt
import json
import csv
```

```python
#import csv file include all US state capital city name

path ="C:/Users/rwang/Desktop/Mentos/usstatecapitalgeocode.csv"
file = pd.read_csv(path)
usstategeodata=pd.DataFrame(file)
usstategeodata.head()
```

# Get 50 restaurants id for each capital city

```python
id=[]
businesscity={}
target_city=usstategeodata["city"]

for city in target_city:
    url="https://api.yelp.com/v3/businesses/search"
    target_search="restaurants"
    params={"term":target_search,'location':city,"limit":50}
    response=requests.get(url,params=params,headers={"Authorization": f"Bearer {ykey}"})
    res=response.json()
    businesscity[city] = [i["id"] for i in res["businesses"]]
    id.append(businesscity[city])
```

# Combine all id into one list

```python
newid=[]
for g in range(49):
    newid+=id[g]
newid
```

```python
newiddata=pd.DataFrame(newid)
newiddata.to_csv("iddata")
```

# Get review count and category information for each restaurant id

```python
review=[]
category=[]
rating=[]
requestcity=[]
lat=[]
lon=[]
for businessid in newid:
    baseurl="https://api.yelp.com/v3/businesses/"
    query_url= f"{baseurl}{businessid}"
    try:
        reviewresponse=requests.get(query_url,headers={"Authorization": f"Bearer {ykey}"})
        reviewres=reviewresponse.json()
        review.append(reviewres["review_count"])
        category.append(reviewres['categories'][0]["alias"])
        rating.append(reviewres["rating"])
        requestcity.append(reviewres['location']["city"])
        lat.append(reviewres['coordinates']["latitude"])
        lon.append(reviewres['coordinates']["longitude"])
    except:
        review.append(np.nan)
        category.append(np.nan)
        rating.append(np.nan)
        requestcity.append(np.nan)
        lat.append(np.nan)
        lon.append(np.nan)
review,category,rating,requestcity,lat,lon
```

# Put id and requsted results into a dataframe

```python
newdata=pd.DataFrame(newid)
newdata
```

```python
newdata["review"]=review
newdata["category"]=category
newdata["rating"]=rating
newdata["city"]=requestcity
newdata["lat"]=lat
newdata["lon"]=lon
newdata.columns=("id","review","category","rating","city","lat","lon")
newdata.sort_values(by="review",ascending=False)
cleandata=newdata.dropna()
cleandata.to_csv("cleandata")
```
