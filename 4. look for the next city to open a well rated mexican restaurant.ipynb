```python
import requests
import pandas as pd
from config import gkey
from scipy import stats
from config import api_key
import numpy as np
from matplotlib import pyplot as plt
import json
import csv
import gmaps
import warnings
warnings.filterwarnings('ignore')
```

#### import analysis data from step2, get city list who doesn't have mexican restaurant

```python
path='C:/Users/rwang/Desktop/Mentos/analysisdata.csv'
file=pd.read_csv(path)
analysisdata=pd.DataFrame(file)
data1=analysisdata[analysisdata["category"]!="mexican"]
data2=data1[data1["category"]!="newmexican" ]
mapdata=data2.drop_duplicates(subset="city",keep="first")
mapdata
```

```python
url1 = "http://api.openweathermap.org/data/2.5/weather?"
query_url1 = f"{url1}appid={api_key}&q="
cities1=mapdata["city"]
temp1=[]
humidity1=[]
windspeed1=[]
cloudness1=[]

for city1 in cities1:
    response = requests.get(query_url1+city1).json()
    try:
        temp1.append(response["main"]['temp'])
        humidity1.append(response["main"]['humidity'])
        windspeed1.append(response["wind"]['speed'])
        cloudness1.append(response["clouds"]['all'])
    except:
        temp1.append(np.nan)
        humidity1.append(np.nan)
        windspeed1.append(np.nan)
        cloudness1.append(np.nan)
```

```python


mapdata['temp']=temp1
mapdata['humidity']=humidity1
mapdata['windspeed']=windspeed1
mapdata['cloudness']=cloudness1
cleanmapdata=mapdata.dropna()
cleanmapdata
```

```python

locations=cleanmapdata[["lat","lon"]]
temp=cleanmapdata["temp"].astype(float)
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations,weights=temp,max_intensity=10,point_radius=2)
fig.add_layer(heat_layer)
fig
```
